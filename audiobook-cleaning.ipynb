```python
df = audio.copy()
import datetime as dt
```

### realease date and price in the their suitable datatype 

```python
# code 
df['releasedate'] = pd.to_datetime(df['releasedate'],dayfirst=True,errors='coerce')
```

```python
df['price'] = df['price'].str.replace(',','') 
df['price'] = df['price'].replace('free','0')
df['price'] = pd.to_numeric(df['price'] , errors='coerce')
```

```python
df.info()
```

## time should be in their correct form

```python
# code
df[['hours','minute']] = df['time'].str.extract(r'(?:(\d+)\s*hr)?.*?(\d+)\s*min')
df[['hours','minute']] = df[['hours','minute']].fillna(0).astype('int')
df['time'] = df['hours']*60 +df['minute']
```

```python
df.drop(columns=['hours','minute'],inplace=True)
```

```python
df.info()
```

```python
df['stars']
```

## star and rating should be in different cols

```python

df['ratings']=df['stars'].str.split().str.get(0).str.replace('Not','not rated').astype('object')
rating_map = {
    "1": "Poor",
    "2": "Fair",
    "3": "Good",
    "4": "Very Good",
    "5": "Excellent",
    "not rated": "Unrated"
}

df['ratings'] = df['ratings'].map(rating_map).fillna(df['ratings'])
```

```python
df['num_ratings'] = df['stars'].str.extract(r'(\d+)\s*ratings')
df['num_ratings'] = pd.to_numeric(df['num_ratings'],errors='coerce').fillna(0).astype('int')
```

```python
df.drop(columns=['stars'],inplace=True)
```

## narrator should be in correct form

```python
df['narrator']=df['narrator'].str.split(':').str.get(1)
```

```python
df['narrator'] = (
    df['narrator']
    .str.replace(r'([A-Z])', r' \1', regex=True)
    .str.strip()
)
```

## author should be in correct form

```python
df['author'] = df['author'].str.split(':').str.get(-1).str.strip()
```

```python
df['author'] = df['author'].str.replace(r'([A-Z])', r' \1', regex=True).str.strip()
```

```python
df['author']
```

```python
df.info()
```

#### Handling Missing Values in `price`
The `price` column had only 0.386% missing values. Since this is negligible, I imputed them using the median (585), which is robust against outliers and preserves the central tendency.

```python
df['price'] = df['price'].fillna(df['price'].median())
```
